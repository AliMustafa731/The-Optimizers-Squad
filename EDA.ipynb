## Import Libraries

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import random
```

## 1. LFW Dataset

### 1.1 Scan the directories

Each directory contains picutres of the same person

```python
dataset_root_path = os.path.join('Data', 'LFW', 'lfw_funneled')

classes_paths = []

# scan all classes directories
for class_dir in os.scandir(dataset_root_path):

    files = os.listdir(class_dir.path)
    classes_paths.append([])

    for file in files:
        classes_paths[-1].append(os.path.join(class_dir.path, file))

total_images_count = np.sum([len(x) for x in classes_paths])
print(f"total number of directories (classes): {len(classes_paths)} people (classes)")
print(f"total number of images of the dataset: {total_images_count} images")
```

therefore, we have 5749 different people (classes)

### 1.2 Analyze the number of photos for  each person (class)

```python
classes_lengths = [len(x) for x in classes_paths]
classes_lengths.sort()
min_images_per_class = classes_lengths[0]
max_images_per_class = classes_lengths[-1]

print(f"min images per person (class): {min_images_per_class}")
print(f"max images per person (class): {max_images_per_class}")
```

### 1.3 Visualize the distribution of images per people

```python
x = np.arange(0, len(classes_paths), 1)
y = np.array([len(x) for x in classes_paths])
```

```python
plt.figure(figsize=(14, 6))
plt.xlabel("person index")
plt.ylabel("number of images")
bar_plot = plt.bar(x, y, 20)
```

We note that there's alot of imbalance in the distribution of images per person.

The problem arises when a person has only one image, this would prevent us from pairing an image from this person with another image from the same person.

This problem can be solved by using Augmentation, as to increase the number of images in the dataset, such that each person (class) has at least two images.

### 1.4 Check images dimensions (width/height)

```python
dims = set()

for clss in classes_paths:
  for path in clss:
    img = plt.imread(path)
    dims.add(img.shape)

dims
```

we notice that all images in the dataset have the same size

### 1.5 Visualize samples from the dataset

```python
# sample 10 random people
people_sample = [plt.imread(x[0]) for x in random.sample(classes_paths, 10)]

plt.figure(figsize=(14, 6))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.axis("off")
    plt.imshow(people_sample[i])
```
